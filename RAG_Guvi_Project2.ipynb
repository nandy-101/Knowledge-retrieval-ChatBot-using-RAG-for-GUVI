```python
# Cell 1: imports
import os
import pandas as pd
from sentence_transformers import SentenceTransformer
import numpy as np

# vector store tools
import faiss

# For LLM calls / generation
from transformers import AutoModelForCausalLM, AutoTokenizer

# your API keys / configuration
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY", "YOUR_OPENAI_KEY_HERE")
# If using OpenAI you will import openai later etc
```

```python
# Cell 2: load your FAQ/blog data
df = pd.read_csv("guvi_faqs.csv")
print("Dataset loaded:", df.shape)
df.head()
```

```python
# Cell 3: (optional) define chunking function if your text is very long
def chunk_text(text, max_words=100):
    words = text.split()
    chunks = []
    for i in range(0, len(words), max_words):
        chunks.append(" ".join(words[i:i+max_words]))
    return chunks

# apply chunking
records = []
for idx, row in df.iterrows():
    chunks = chunk_text(row["chunk_text"], max_words=50)
    for c in chunks:
        records.append({"id": row["id"], "source": row["source"], "chunk": c})
df_chunks = pd.DataFrame(records)
print("After chunking:", df_chunks.shape)
df_chunks.head()
```

```python
# Cell 4: embedding generation
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
embeddings = model.encode(df_chunks["chunk"].tolist(), show_progress_bar=True)
print("Embeddings shape:", embeddings.shape)
```

```python
# Cell 5: build FAISS index
d = embeddings.shape[1]
faiss_index = faiss.IndexFlatL2(d)  # L2 distance
faiss_index.add(np.array(embeddings))
print("Total vectors indexed:", faiss_index.ntotal)
```

```python
# Cell 6: retrieval function to get top-k relevant chunks
def retrieve(query, top_k=3):
    q_emb = model.encode([query])
    D, I = faiss_index.search(q_emb, top_k)
    results = []
    for idx in I[0]:
        results.append(df_chunks.iloc[idx]["chunk"])
    return results

# test retrieval
test_q = "How can I enroll in a GUVI course?"
print("Retrieved chunks:", retrieve(test_q, top_k=2))
```

```python
# Cell 7: load or call LLM for answer generation
tokenizer = AutoTokenizer.from_pretrained("gpt2")  # example: using GPT-2 small
model_gen = AutoModelForCausalLM.from_pretrained("gpt2")

def generate_answer(query, context_chunks):
    # combine context + query into prompt
    prompt = "Context: " + "\n".join(context_chunks) + "\n\nQuestion: " + query + "\nAnswer:"
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model_gen.generate(inputs["input_ids"], max_length=inputs["input_ids"].shape[1]+50,
                                 num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return answer[len(prompt):]

# test generation
chunks = retrieve(test_q, top_k=2)
print("Generated answer:", generate_answer(test_q, chunks))
```

```python
# Cell 8: simple chat interface in notebook
print("Chat with the GUVI RAG Bot (type ‘exit’ to quit)\n")
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        break
    chunks = retrieve(user_input, top_k=3)
    answer = generate_answer(user_input, chunks)
    print("Bot:", answer)
```

```python
# Cell 9: create Streamlit app
%%writefile app.py
import streamlit as st
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from transformers import AutoTokenizer, AutoModelForCausalLM
import pandas as pd

# load data, embeddings, index
df_chunks = pd.read_csv("guvi_faqs_chunks.csv")  # you will export df_chunks
embeddings = np.load("embeddings.npy")
d = embeddings.shape[1]
faiss_index = faiss.IndexFlatL2(d)
faiss_index.add(embeddings)

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model_gen = AutoModelForCausalLM.from_pretrained("gpt2")

def retrieve(query, top_k=3):
    q_emb = model.encode([query])
    D, I = faiss_index.search(q_emb, top_k)
    return [df_chunks.iloc[idx]["chunk"] for idx in I[0]]

def generate_answer(query, chunks):
    prompt = "Context: " + "\n".join(chunks) + "\n\nQuestion: " + query + "\nAnswer:"
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model_gen.generate(inputs["input_ids"], max_length=inputs["input_ids"].shape[1]+50,
                                 num_return_sequences=1, pad_token_id=tokenizer.eos_token_id)
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return answer[len(prompt):]

st.title("GUVI Knowledge Chatbot")
user_input = st.text_input("Ask me anything about GUVI courses, blogs, docs:")
if user_input:
    chunks = retrieve(user_input, top_k=3)
    st.write("**Retrieved context:**")
    for c in chunks:
        st.write("-", c)
    answer = generate_answer(user_input, chunks)
    st.write("**Answer:**", answer)
```
